Question 1
 plot a line plot between a and b:
a=np.arange(40,50)
b=np.arange(50,60)



```python
# Exercie 1:
a= np.arange(40,50)
b= np.arange(50,60)
plt.plot(a,b,'r*--')
plt.title('line plot', color='blue')
plt.xlabel('x_axis',color='green')
plt.ylabel('y_axix',color='green')
plt.show()
```

Question 2
Plot a line plot showing the sales trend in company 1 and 2:

days = [1,2,3,4,5,6,7] #days of d week

sales_1 = [160,150,140,145,175,165,180] #sales of company1

sales_2 = [70,90,160,150,140,145,175]  #sales of company2


```python
#Exescise 2:
days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175] #sales of company2
plt.subplot(2,1,1)
plt.plot(days,sales_1)
plt.xlabel('days')
plt.ylabel('sales_1')
plt.subplot(2,1,2)
plt.plot(days,sales_2)
plt.xlabel('days')
plt.ylabel('slaes_2')
plt.show()
```

```python
#Exercise 2:
days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175] #sales of company2
plt.plot(days,sales_1,label='sale_1')
plt.plot(days,sales_2,label='sale_2')
plt.xlabel('days')
plt.ylabel('sales')
plt.legend(loc='upper left')
plt.show()
```
